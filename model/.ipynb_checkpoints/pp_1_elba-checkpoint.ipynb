## Rename Data

```python
data = data.rename(columns={"Ereignis": "Scenario",'Nummer':'experiment_number'})

# Rename labels
data = data.replace({'Scenario': {'Unknown': 'Background',
                                  'Deo': 'Deodorant',
                                 'Ethanol': 'Ethanol',
                                 'Haarspray': 'Hairspray',
                                 'Holz': 'Wood',
                                 'Kabel':'Cable',
                                 'Kerzen':'Candles',
                                 'Lunten':'Lunts'                                   
                                 }})

# Rename sensor ID
data.Sensor_ID = data.Sensor_ID.str.replace('Sensorknoten', 'sensornode')
```

## Define Datetime index

```python
# Sort df by Date
data = data.sort_values(by=['Date'])

# parse Date Column as DateTime
#data['Date'] = pd.to_datetime(data['Date'])

# Set Date as Index Column
data = data.set_index('Date') # data = data.set_index(pd.to_datetime(data['Date']))

# resample to reduce df Rows
#data.resample('2min').mean()
```

#### Drop Cable Experiment with Kurzschluss

```python
# mask_drop = (data.index >='2022-07-07 14:18:50') & (data.index <='2022-07-07 14:50:00') # Zwei mal Kurzschluss --> Versuchsabbruch
mask_drop = (data.Scenario == 'Cable') & (data.experiment_number == 'E2')
rows_drop = data[mask_drop].index
data = data.drop(index=rows_drop)
```

#### Drop CO2 Values were threshold is exceeded because of humans

```python
#data.loc[(data['CO2_Room']> 1050) & (data.index <= '2022-07-04 18:00:00'),'CO2_Room']=np.nan
```

#### Replace UV Signal from wood scenario with zero

```python
data.loc[(data['UV_Room']>=2) & (data.Scenario == 'Wood'),'UV_Room']= 0 #np.nan
```
